# 어떤 선수를 보강해야할까요?

```python
df = data[(data['Club'] == 'Chelsea')|(data['Club']=='Manchester City')]
```

```python
df['Club'].unique()
```

```python
# 이적료 비교
df['Value'].head()
```

```python
import warnings

warnings.filterwarnings(action= 'ignore')
```

```python
df['Value'] = df['Value'].str.replace('M', '000000')
df['Value'] = df['Value'].str.replace('K', '000')
```

```python
df['Value']
```

```python
df['Value']=df['Value'].str.slice(1,)
```

```python
df['Value']=df['Value'].str.replace('.', '')
```

```python
df['Value'] = df['Value'].astype(int)
```

```python
df.head()
```

```python
ch = df[df['Club']=='Chelsea']
mc = df[df['Club']=='Manchester City']
```

```python
mc.head()
```

```python
gk_list = ['GK']
cb_list = ['CB', "LCB", 'RCB', 'RB', 'LB']
mf_list = ['RCM', 'LCM', 'RDM', 'CDM', 'CM', 'RM', 'CAM']
st_list = ['ST', 'LW', 'RW', 'CF']
```

```python
st_count = 2
mf_count = 4
cb_count = 4
gk_count = 1

ch_id=[]

for index in ch.index:
    if ch['Position'][index] in gk_list :
      if gk_count != 0 :
          ch_id.append(ch['ID'][index])
          gk_count -= 1
    elif ch['Position'][index] in cb_list :
      if cb_count != 0 :
          ch['Position'][index] = 'CB'
          ch_id.append(ch['ID'][index])
          cb_count -= 1
    elif ch['Position'][index] in mf_list :
      if mf_count != 0 :
          ch['Position'][index] = 'MF'  
          ch_id.append(ch['ID'][index])
          mf_count -= 1
    else :
      if st_count != 0 :
          ch['Position'][index] = 'ST'
          ch_id.append(ch['ID'][index])
          st_count -= 1
```

```python
ch_id
```

```python
ch = ch[ch['ID'].isin(ch_id)]
```

```python
st_count = 2
mf_count = 4
cb_count = 4
gk_count = 1

mc_id=[]

for index in mc.index:
    if mc['Position'][index] in gk_list :
      if gk_count != 0 :
          mc_id.append(mc['ID'][index])
          gk_count -= 1
    elif mc['Position'][index] in cb_list :
      if cb_count != 0 :
          mc['Position'][index] = 'CB'
          mc_id.append(mc['ID'][index])
          cb_count -= 1
    elif mc['Position'][index] in mf_list :
      if mf_count != 0 :
          mc['Position'][index] = 'MF'        
          mc_id.append(mc['ID'][index])
          mf_count -= 1
    else :
      if st_count != 0 :
          mc['Position'][index] = 'ST'
          mc_id.append(mc['ID'][index])
          st_count -= 1
```

```python
mc_id
```

```python
mc = mc[mc['ID'].isin(mc_id)]
```

```python
mc.head()
```

```python
df = pd.concat([ch, mc])
```

```python
df
```

```python
sns.boxplot(data = df, x='Position', y= 'Overall', hue='Club')
```

```python
sns.boxplot(data = df, x='Position', y='Value', hue='Club')
```

```python
f, ax = plt.subplots(2,4,figsize=(20,10))

vs_list = ['Age', 'Overall', 'Potential', 'Weak Foot']

for i in  range(8):
  if i<4:

    colors = ['firebrick' if x > market[market['Position']=='CB'][:13][vs_list[i]].mean() else 'gray' for x in market[market['Position']=='CB'][:13][vs_list[i]]]
    sns.barplot(x=vs_list[i], y='Name', data=market[market['Position']=='CB'][:13], ax=ax[i//4, i%4], palette = colors)

    ax[i//4, i%4].axvline(market[market['Position']=='CB'][:13][vs_list[i]].mean(), ls='--')
  else:
    sns.barplot(x = vs_list[i%4], y='Name', data= market[market['Position']=='RM'][:13], ax=ax[i//4, i%4])

    ax[i//4, i%4].axvline(market[market['Position']=='RM'][:13][vs_list[i%4]].mean(), ls='--')
```
