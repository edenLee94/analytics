2번째 page에 대한 확인 과정을 통해서 page의 흐름이 어떻게 가는지 확인한다.

```python
html =req.text
soup = BeautifulSoup(html, 'lxml')
title_list = soup.find_all('h4', 'titles')
list(map(lambda x : x.a.text, title_list))
```

```python
title_info = list()
for page in range(1,10):
# for page in page_num:
    url = 'https://www.thekpm.com/news/articleList.html?page=' + str(page) +'&total=1526&sc_section_code=S1N2&sc_sub_section_code=&sc_serial_code=&sc_area=&sc_level=&sc_article_type=&sc_view_level=&sc_sdate=&sc_edate=&sc_serial_number=&sc_word=&sc_word2=&sc_andor=&box_idxno=&sc_multi_code=&sc_is_image=&sc_is_movie=&sc_user_name=&sc_order_by=E'
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html.parser')
    title_list = soup.find_all('h4', 'titles')

    for title in title_list:
      title_info.append(title.text)

    page += 1
    time.sleep(1)
```

```python
# 해당 page의 대표 topic에 대한 정보를 넣어준다
topic_info = 'smartfarm'
farm_df = pd.DataFrame(title_info, columns = ['title'])
farm_df['topic']= topic_info
farm_df
```

```python
title_info2 = list()
for page2 in range(1,10):
# for page in page_num:
    urls = 'https://www.thekpm.com/news/articleList.html?page=' + str(page2) +'&total=1526&sc_section_code=S1N2&sc_sub_section_code=&sc_serial_code=&sc_area=&sc_level=&sc_article_type=&sc_view_level=&sc_sdate=&sc_edate=&sc_serial_number=&sc_word=&sc_word2=&sc_andor=&box_idxno=&sc_multi_code=&sc_is_image=&sc_is_movie=&sc_user_name=&sc_order_by=E'
    reqs = requests.get(urls)
    soup2 = BeautifulSoup(reqs.text, 'html.parser')
    title_list2 = soup2.find_all('h4', 'titles')

    for title2 in title_list2:
      title_info2.append(title2.text)

    page2 += 1
    time.sleep(1)
```

```python
# 해당 page의 대표 topic에 대한 정보를 넣어준다
econo_df = pd.DataFrame(title_info2, columns = ['title'])
econo_df['topic']= 'economy'
econo_df
```

```python
news_df = pd.concat([farm_df,econo_df])
```

```python
news_df
```

```python
news_df = news_df.reset_index(drop=True)
```

```python
news_df.drop_duplicates('title', inplace=True) # 중복 제거 
```

```python
news_df
```

```python
news_df["title"] = news_df["title"].str.replace(pat=r'[^\w]',repl=r' ',regex=True) # 특수문자 제거 
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
ko_vectorizer = TfidfVectorizer(min_df = 3, ngram_range = (1,3), tokenizer = okt.nouns)
```

```python
features = ko_vectorizer.fit_transform(news_df['title'])
```

```python
columns = ko_vectorizer.get_feature_names_out()
pd.DataFrame(features.todense(), columns = columns)
```

```python
cos_sim = cosine_similarity(features, features, dense_output=True)
```

```python
pd.DataFrame(cos_sim)
```

```python
index = pd.DataFrame(cos_sim)[5].sort_values(ascending = False)[:11].index
```
